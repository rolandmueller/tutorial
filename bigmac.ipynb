```python
import pandas as pd
```

```python
df = pd.read_csv("bigmac.csv")
```

```python
(
    df
    .assign(date=lambda d: pd.to_datetime(d['date']))
    .set_index('date')
    .rename(columns={'name': 'country', 'currency_code': 'currency'})
    .assign(dollar_price=lambda d: d['local_price']/d['dollar_ex'])
    .assign(currency_value=lambda d: d['dollar_price'] / d.query('country=="United States"')['dollar_price'])
    .loc[lambda d: d.index.max()]
    .sort_values('currency_value', ascending=False)
    .head(10)
)
```
